# Quais os top 15 na área de reprodutibilidade cientifica?

Ótima pergunta. Em **reprodutibilidade científica**, não existe um ranking único oficial, mas este é um **Top 15 amplamente aceito na prática** (especialmente em pesquisa computacional):

1.  **Git** — versionamento de código e histórico auditável. ([Git](https://git-scm.com/book/ms/v2/Getting-Started-About-Version-Control?utm_source=chatgpt.com "1.1 Getting Started - About Version Control"))
2.  **GitHub / GitLab** — colaboração, revisão e integração com CI/CD (base de workflows reproduzíveis). ([GitHub Docs](https://docs.github.com/en/get-started/using-git/about-git?utm_source=chatgpt.com "About Git"))
3.  **Jupyter Notebook / JupyterLab** — notebooks para computação interativa e documentação do processo analítico. ([Jupyter Documentation](https://docs.jupyter.org/?utm_source=chatgpt.com "Jupyter Documentation - Jupyter Notebook"))
4.  **Quarto** — publicação técnico-científica reprodutível (HTML/PDF/Word) com Python/R/Julia. ([Quarto](https://quarto.org/?utm_source=chatgpt.com "Quarto"))
5.  **Snakemake** — orquestração de pipelines reprodutíveis e escaláveis. ([Snakemake](https://snakemake.readthedocs.io/?utm_source=chatgpt.com "Snakemake | Snakemake 9.16.2 documentation"))
6.  **Nextflow** — workflows portáveis/reprodutíveis para HPC e nuvem. ([Nextflow](https://www.nextflow.io/docs/latest/?utm_source=chatgpt.com "Nextflow documentation"))
7.  **Docker** — empacotamento de ambiente para execução consistente. ([Docker](https://www.docker.com/blog/docker-for-devops/?utm_source=chatgpt.com "Exploring Docker for DevOps: What It Is and How It Works"))
8.  **Apptainer (Singularity)** — containers portáveis muito usados em HPC acadêmico. ([Apptainer](https://apptainer.org/?utm_source=chatgpt.com "Apptainer - Portable, Reproducible Containers"))
9.  **Conda** — gestão de ambientes e dependências com exportação/compartilhamento. ([Conda Documentation](https://docs.conda.io/docs/user-guide/tasks/manage-environments.html?utm_source=chatgpt.com "Managing environments — conda 26.1.1.dev6 ..."))
10. **DVC (Data Version Control)** — versionamento de dados/modelos e pipelines de ML reproduzíveis. ([Data Version Control · DVC](https://doc.dvc.org/user-guide?utm_source=chatgpt.com "User Guide | Data Version Control · DVC"))
11. **MLflow** — rastreio e reprodução de experimentos de ML. ([MLflow](https://mlflow.org/classical-ml/experiment-tracking?utm_source=chatgpt.com "Experiment tracking"))
12. **OSF (Open Science Framework)** — gestão de projeto científico, colaboração e transparência do ciclo de pesquisa. ([OSF Support](https://help.osf.io/article/353-welcome-to-projects?utm_source=chatgpt.com "Welcome to OSF Projects"))
13. **Zenodo** — depósito/arquivamento com DOI para dados, software e artefatos. ([Zenodo](https://about.zenodo.org/?utm_source=chatgpt.com "Zenodo: About"))
14. **DataLad** — gestão distribuída de dados com proveniência e foco em FAIR/reprodutibilidade. ([DataLad](https://www.datalad.org/?utm_source=chatgpt.com "DataLad – DataLad"))
15. **Jupyter Book** — documentação/publicação reprodutível e interativa baseada no ecossistema Jupyter. ([Jupyter Book](https://jupyterbook.org/?utm_source=chatgpt.com "Jupyter Book"))

# Liste os 15 principais softwares utilizados em Open Science

Excelente recorte. Em **Open Science**, os “principais” costumam ser os que cobrem: planejamento/pré-registro, colaboração aberta, repositórios, versionamento, workflows reprodutíveis e publicação aberta.

## Top 15 softwares/plataformas usados em Open Science

1.  **Open Science Framework (OSF)** — gestão de projeto, colaboração, compartilhamento e pré-registro. ([Center for Open Science](https://www.cos.io/products/osf?utm_source=chatgpt.com "The Open Science Framework"))
2.  **Zenodo** — repositório aberto para dados/software/publicações com DOI. ([Zenodo](https://about.zenodo.org/?utm_source=chatgpt.com "Zenodo: About"))
3.  **Dataverse** — software open source para repositórios de dados de pesquisa. ([Dataverse](https://dataverse.org/about?utm_source=chatgpt.com "About"))
4.  **Galaxy** — plataforma aberta para análises científicas reprodutíveis por workflow (muito forte em bioinfo). ([Galaxy Project](https://galaxyproject.org/?utm_source=chatgpt.com "Galaxy Project"))
5.  **Jupyter Notebook / JupyterLab** — notebooks reprodutíveis (código + narrativa + resultado).
6.  **Quarto** — publicação científica reprodutível (HTML/PDF/Word) com notebooks.
7.  **RStudio + R Markdown** — pipeline reprodutível para análise e relatório científico.
8.  **Git** — controle de versão rastreável para transparência metodológica.
9.  **GitHub** — colaboração aberta, revisão, issues e integração com publicação de artefatos.
10. **GitLab** — alternativa forte ao GitHub com CI/CD integrado para workflows científicos.
11. **Docker** — empacotamento de ambiente para repetibilidade computacional.
12. **Apptainer (Singularity)** — containerização focada em HPC acadêmico.
13. **Snakemake** — orquestração declarativa de pipelines reprodutíveis.
14. **Nextflow** — workflows científicos portáveis (local, cluster e cloud).
15. **DVC (Data Version Control)** — versionamento de dados/modelos e pipelines de experimentos.
