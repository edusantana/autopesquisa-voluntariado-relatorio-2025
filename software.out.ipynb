## CosmoWare

Quando eu estava acessando os sites das ICs, especialmente o site da Reaprendentia, eu percebi que havia artes demonstrando as equipes, no entanto essas artes estavam desatualizadas. Esta foi minha motivação para criar uma extensão para o Google Chrome que batizei com o nome de `Cosmoware`.

Desenvolvi a funcionalidade de gerar diagramas das equipes a partir das informações cadastradas no ICNET (ver @fig-cosmoware-organograma).

{#fig-cosmoware-organograma }

Posteriormente adicionei outra funcionalidade de exportar os dados do ICNET em planilhas compatíveis com o RStudio, ver @fig-cosmoware.

{#fig-cosmoware }

> **Código Pessoal de Cosmoética escrito há 6 nos**
>
> A instalação do Cosmoware pode ser realizada a partir da [loja do Google Chrome através deste link](https://chromewebstore.google.com/detail/conscienciologia-tools-co/mamilhfnkleimaphdjpkinkekdbkbacl).

## Instant Data Scraper

Para a extração de dados do ICNet (ver @fig-atividades) copiar e colar no Excel é suficiente.

Extrações mais complexas podem ser realizadas com a extensão para o Google Chrome chamada Instant Data Scraper, criada por @instantDataScraper. A @fig-instantdata mostra uma extração de uma lista de pedidos de compras na Megastore.

{#fig-instantdata }

## Track & Graph

O Track & Graph (@fig-trackandgraph) é um aplicativo para Android, @trackandgraph, que permite o registro de dados pessoas e exportação dos dados para planilhas. Pode ser utilizado para registro de tentativas de EV, de peso, sinaléticas ou qualquer outras variáveis que forem configuradas.

{#fig-trackandgraph }

## RStudio

O RStudio mantido por @rstudio é uma IDE (Ambiente de Desenvolvimento Integrado) que permite a edição dos documentos e agrupa diversas funcionalidades para gerenciar o projeto. Ela é uma ferramenta de referência para a ciência de dados de código aberto.

## Quarto

O Quarto mantido por @quarto é um sistema de código aberto para escrita e publicação técnica e científica. Ela é uma ferramenta de comandos de linhas para gerenciar e publicar projetos, integrada nas versões recentes do RStudio.
